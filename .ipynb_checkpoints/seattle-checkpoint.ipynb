```python
import urllib2
import json
import pandas
import seaborn
import matplotlib.pyplot as plt
seaborn.set()
```

```python
# adding some supplementary weather data: http://www.ncdc.noaa.gov/, following blogpost above
weather = pandas.read_csv("seattle_NOAA.csv",parse_dates=True,index_col='DATE')
```

```python
# getting the data
request = urllib2.urlopen("https://data.seattle.gov/resource/4xy5-26gy.json")
data = pandas.DataFrame(json.loads(request.read()))
data.head()
```

```python
# some manipulations
data.index = pandas.DatetimeIndex(data['date'])
data = data.drop(["date"],axis=1)
data['crossings'] = data['fremont_bridge_sb'].apply(int) + data['fremont_bridge_nb'].apply(int)
data = data.drop(["fremont_bridge_nb","fremont_bridge_sb"],axis=1)
#data = data.drop(["fremont_bridge_sb","fremont_bridge_nb"])
```

```python
# above shows that the dataset is totally non-sorted
data = data.sort_index(0)
data.head()
```

```python
# now let's aggregate hourly
data_hourly = data.resample("1h").sum().dropna()
data_daily  = data.resample("1d").sum().dropna()
data_weekly = data.resample("1w").sum().dropna()
data_hourly.head()
```

```python
fig, ax = plt.subplots(ncols=1,nrows=1,figsize=(8,6))
data_weekly.plot(ax=ax,label='weekly',title='weekly crossings Fremont Seattle, WA')
```

```python
pandas.rolling_mean(data.crossings,window='30d').plot(title='rolling average (monthly)')
```

```python
# daylight modeling: https://jakevdp.github.io/blog/2014/06/10/is-seattle-really-seeing-an-uptick-in-cycling/
def hours_of_daylight(date, axis=23.44, latitude=47.61):
    """Compute the hours of daylight for the given date"""
    diff = date - pandas.datetime(2000, 12, 21)
    day = diff.total_seconds() / 24. / 3600
    day %= 365.25
    m = 1. - np.tan(np.radians(latitude)) * np.tan(np.radians(axis) * np.cos(day * np.pi / 182.625))
    m = max(0, min(m, 2))
    return 24. * np.degrees(np.arccos(1 - m)) / 180.

# add this to our weekly data
data_weekly['daylight'] = map(hours_of_daylight, data_weekly.index)
data_daily['daylight'] = map(hours_of_daylight, data_daily.index)
```

```python
data_weekly.plot(x='daylight',y='crossings',kind='scatter')
```

```python
# spearman rank correlation test
data_weekly.corr("spearman")
```

```python
from sklearn.linear_model import LinearRegression
# the underlying model here is really a *linear* model
```

```python
X = data_weekly[['daylight']].to_dense() # pdf?
y = data_weekly['crossings']
clf = LinearRegression(fit_intercept=True).fit(X,y)
print(clf.coef_[0])

data_weekly['daylight_trend'] = clf.predict(X)
data_weekly['daylight_corrected'] = data_weekly['crossings'] - data_weekly['daylight_trend']
```

```python
data_weekly.crossings.plot()
```

```python
xfit = np.linspace(7, 17)
yfit = clf.predict(xfit[:, None])
plt.scatter(data_weekly['daylight'], data_weekly['crossings'])
plt.plot(xfit, yfit, '-k')
plt.title("Bicycle traffic through the year")
plt.xlabel('daylight hours')
plt.ylabel('weekly bicycle traffic');
```

```python
trend = clf.predict(data_weekly[['daylight']].as_matrix())
plt.scatter(data_weekly['daylight'], data_weekly['crossings'] - trend + np.mean(trend))
plt.plot(xfit, np.mean(trend) + 0 * yfit, '-k')
plt.title("weekly traffic (detrended)")
plt.xlabel('daylight hours')
plt.ylabel('adjusted weekly count');
```

```python
data_weekly[['crossings','daylight_trend']].plot()
```

```python
data_weekly['daylight_corrected'].plot()
rms = np.std(data_weekly['daylight_corrected'])
plt.ylabel("adjusted total weekly riders")
print("root-mean-square about trend: {0:.0f} riders".format(rms))
```

```python
# accounting for day of week
days = ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun']
data_daily['dayofweek'] = data_daily['crossings'].index.dayofweek

grouped = data_daily.groupby('dayofweek')['crossings'].mean()
grouped.index = days
```

```python
grouped.plot(title="Avg. Crossings per weekday")
```

```python
# simultaneously detrend using daylight & day stuff

# Add one-hot indicators of weekday
for i in range(7):
    data_daily[days[i]] = (data_daily.index.dayofweek == i).astype(float)

# de-trend on days of the week and daylight together
## the *implicit* assumption is that there's a linear relation between the day of the week / daylight & crossings

X = data_daily[days + ['daylight']] 
y = data_daily['crossings']
clf = LinearRegression().fit(X, y)
data_daily['dayofweek_trend'] = clf.predict(X)
data_daily[['crossings', 'dayofweek_trend']].plot();
```

```python
# look at detrended data
data_daily['dayofweek_corrected'] = (data_daily['crossings'] - data_daily['dayofweek_trend'] + data_daily['dayofweek_trend'].mean())
print("rms = {0:.0f}".format(np.std(data_daily['dayofweek_corrected'])))
data_daily['dayofweek_corrected'].plot();
```

```python
# join the weather data to our daily data
data_daily = data_daily.join(weather[['TMIN','TMAX','PRCP']])
data_daily['TMIN2'] = data_daily['TMIN']**2
data_daily['TMAX2'] = data_daily['TMAX']**2
#data_daily['PRCP2'] = data_daily.PRCP**2

columns = days + ['daylight', 'TMIN', 'TMAX', 'PRCP']
columns += ['TMIN2','TMAX2']
```

Multivariate Linear Regression

$y_{ik} = b_{0k} + \sum_{j=1}^{p}{b_{jk}x_{ij}} + e_{ik}$

* $y_{ik}$ is the k-th real-valued response for the i-th observation
* $b_{0k}$ is the regression intercept for k-th response
* $b_{jk}$ is the j-th predictors regression slope for k-th response
* $x_{ij}$ is the j-th predictor for the i-th observation
* $e_{ik}$ is multivariate Gaussian error vector


```python
X = data_daily[columns]
# again, the multivariate regression makes the assumption that all support variables are linear with respect to y.
y = data_daily['crossings']
clf = LinearRegression().fit(X, y)
data_daily['overall_trend'] = clf.predict(X)

# Plot the overall trend
data_daily[['crossings', 'overall_trend']].plot()
plt.ylabel('Daily bicycle traffic');
```

```python
data_daily['overall_corrected'] = data_daily['crossings'] - data_daily['overall_trend'] + data_daily['overall_trend'].mean()
print("rms = {0:.0f}".format(np.std(data_daily['overall_corrected'])))
```

```python
data_daily.overall_corrected.rolling("100d").mean().plot(title='100 days rolling average',label='corrected')
data_daily.crossings.rolling("100d").mean().plot(label='raw')
plt.gca().legend()
plt.gca().set_ylabel("Normalized Bike Traffic")
```
